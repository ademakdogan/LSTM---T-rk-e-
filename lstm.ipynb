```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import math
import sklearn
import sklearn.preprocessing
import datetime
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from keras.optimizers import adam
```

```python
data = pd.read_csv("/Users/.../prices-split-adjusted.csv")

#Uniqe symbol
unique_values = data["symbol"].unique()
unique_freq = data.groupby('symbol').count()

df  = data.loc[data["symbol"] == "EQIX"]
```

```python
def normalize_data(df):
    min_max_scaler = sklearn.preprocessing.MinMaxScaler()
    df['open'] = min_max_scaler.fit_transform(df.open.values.reshape(-1,1))
    df['high'] = min_max_scaler.fit_transform(df.high.values.reshape(-1,1))
    df['low'] = min_max_scaler.fit_transform(df.low.values.reshape(-1,1))
    df['close'] = min_max_scaler.fit_transform(df['close'].values.reshape(-1,1))
    return df
```

```python
df = normalize_data(df)
df = df.reset_index(drop = True)
df = df.drop(["date", "symbol","volume"], axis = 1)

training_data = df.iloc[:-200]
test_data = df.iloc[-200:]
df2 = training_data[10 : 20]
training_data = training_data.as_matrix()
X_train = []
y_train = []

for i in range(20, training_data.shape[0]):
    X_train.append(training_data[i-20:i])
    y_train.append(training_data[i, 0])


X_train , y_train = np.array(X_train), np.array(y_train)
```

```python
#LSTM MODEL 
regressor = Sequential()
regressor.add(LSTM(units = 100, activation = "relu",
                   return_sequences = True, input_shape = (X_train.shape[1],X_train.shape[2])))

regressor.add(LSTM(units = 100, activation = "relu", return_sequences = True))

regressor.add(LSTM(units = 120, activation = "relu", return_sequences = True))
regressor.add(Dropout(0.5))

regressor.add(LSTM(units = 120, activation = "relu"))
regressor.add(Dense(units = 1))
regressor.summary()

regressor.compile(optimizer = "adam", loss = "mean_squared_error")
regressor.fit(X_train, y_train, epochs = 500, batch_size = 80)
```

```python
data_training = df.iloc[:-200]
past_20_days = data_training.tail(20)
df_test = past_20_days.append(test_data, ignore_index = True)
test_data = df_test.as_matrix()

X_test = []
y_test = []

for i in range(20, test_data.shape[0]):
    X_test.append(test_data[i-20:i])
    y_test.append(test_data[i,0])

X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
y_pred = regressor.predict(X_test)
```
